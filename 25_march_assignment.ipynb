#### Q13. How does the type of app affect its price? Create a bar chart to compare average prices by app type. 

- There are only two types of apps:
  1. Free
  2. Paid


```python
import pandas as pd
import matplotlib.pyplot as plt

# Change the Price data to float value
df['Price'] = df['Price'].str.split('$').str[1].astype(float)

# Group the dataset by app type and calculate the average price
grouped = df.groupby('Type')['Price'].mean()
grouped.fillna(0, inplace=True)

# Create a bar chart
plt.bar(grouped.index, grouped.values)
plt.xlabel('Type')
plt.ylabel('Average Price')
plt.title('Average Prices by App Type')

# Show the plot
plt.show()
```

#### Q14. What are the top 10 most popular apps in the dataset? Create a frequency table to identify the apps with the highest number of installs.

- The table is below with the top 10 apps with the highest no of installs.

```python
df['Installs'] = df['Installs'].str.replace(',', '')
df['Installs'] = df['Installs'].str.replace('+', '').astype(int)

freq_table = df.sort_values(by=['Installs'], ascending=False)
freq_table[['App', 'Installs']].reset_index(drop=True).head(10)
```
